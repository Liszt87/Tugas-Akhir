# Convert to Pickle Extension

```python
def conv2pickle():
    for i in range(len(myList)):
        main = myList[i][['category', 'imageURLHighRes']]
        main[(main['category'].str.len() != 0) 
             & (myList[i]['imageURLHighRes'].str.len() != 0)].to_pickle('df' + str(i) + '.pickle')

conv2pickle()
```

# Combine All of Pickle File

```python
def df_combine():
    init_pickle = pd.DataFrame(columns = ['category','imageURLHighRes'])
    for i in range(29):
        pick = pd.read_pickle('df'+str(i)+'.pickle')
        init_pickle = pd.concat([init_pickle, pick])
        
    #reset index & save final pickle data frame
    final_pickle =  init_pickle.reset_index(drop=True).to_pickle('df_final.pickle')
        
df_combine()
```
