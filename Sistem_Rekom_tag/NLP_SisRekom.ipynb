# Text Processing
## Data Cleaning

```python
#attribute Category 
def data_clean_category(attribute):
    
    for i in range(len(df_main[attribute])):
        if '</span></span></span>' not in df_main[attribute][i]:
            continue
        else:
            df_main[attribute][i].remove('</span></span></span>')
    
    return df_main[attribute]
    
df_main['category'] = data_clean_category(attribute = 'category')
df_main['category']
```

```python
#attribute Description

CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext

def data_clean_desc(attribute):
    for i in range(len(df_main[attribute])):
        if not df_main[attribute][i]:
            continue
        elif len(df_main[attribute][i]) > 1:
            for j in range(len(df_main[attribute][i])):
                df_main[attribute][i][j] = cleanhtml(df_main[attribute][i][j])
                df_main[attribute][i][j] = re.sub(r'   ', '', df_main[attribute][i][j])
        else:
            df_main[attribute][i][0] = cleanhtml(df_main[attribute][i][0])
            df_main[attribute][i][0] = re.sub(r'   ', '', df_main[attribute][i][0])
    return df_main[attribute]
            
df_main['description'] = data_clean_desc(attribute = 'description')

df_main['description']
```

```python
#attribute brand
def data_clean_brand(attribute):
    for i in range(len(df_main[attribute])):
        df_main[attribute][i] = re.sub(r'by\n    \n    ', '', df_main[attribute][i]) 
    
    return df_main[attribute]

df_main['brand'] = data_clean_brand(attribute = 'brand')

df_main['brand']
```

## Normalisasi Teks

```python
import contractions

def norm_text_desc(attribute):
    for i in range(len(df_main[attribute])):
        if not df_main[attribute][i]:
            continue
        elif len(df_main[attribute][i]) > 1:
            for j in range(len(df_main[attribute][i])):
                expanded_words = []  
                for word in df_main[attribute][i][j].split():
                    expanded_words.append(contractions.fix(word))
                    
                df_main[attribute][i][j] = ' '.join(expanded_words)
        else:
            expanded_words = []  
            for word in df_main[attribute][i][0].split():
                expanded_words.append(contractions.fix(word))
            
            df_main[attribute][i][0] = ' '.join(expanded_words)
    return df_main[attribute]

df_main['description'] = norm_text_desc(attribute = 'description')
```

## Menghilangkan Tanda Baca

```python
from num2words import num2words
#Num2words
def _conv_num(match):
    return num2words(match.group())

def numbers_to_words(text):
    return re.sub(r'\b\d+\b', _conv_num, text)
```

```python
#Menghilangkan tanda baca 
def clean_text(Text):
    Text = re.sub(r'https://[A-Za-z0-9./]+','',Text) # menghapus url  
    Text = numbers_to_words(Text) # menghapus angka
    Text = ''.join([char for char in Text if char not in string.punctuation]) # menghapus tanda baca
    Text = re.sub (r',','',Text) # menghapus koma
    return Text
```

```python
def deleting_punc_in_the_list(attribute):
    for i in range(len(df_main[attribute])):
        if not df_main[attribute][i]:
            continue
        elif len(df_main[attribute][i]) > 1:
            for j in range(len(df_main[attribute][i])):
                df_main[attribute][i][j] = clean_text(df_main[attribute][i][j])
        else:
            df_main[attribute][i][0] = clean_text(df_main[attribute][i][0])
    return df_main[attribute]

df_main['description'] = deleting_punc_in_the_list(attribute = 'description')
df_main['feature'] = deleting_punc_in_the_list(attribute = 'feature')

def deleting_punc_text(attribute):
    for i in range(len(df_main[attribute])):
        df_main[attribute][i] = clean_text(df_main[attribute][i])
    
    return df_main[attribute]

df_main['title'] = deleting_punc_text(attribute = 'title')
df_main['brand'] = deleting_punc_text(attribute = 'brand')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances

class Tag_Recommender_System:
    def __init__(self, data, konten_kol):
        self.df = data
        self.konten_kol = konten_kol
        self.encoder = None
        self.bank = None
        
    def fit(self):
        self.encoder = TfidfVectorizer(stop_words = 'english', tokenizer = word_tokenize)
        self.bank = self.encoder.fit_transform(self.df[self.konten_kol])
        
    def rekomendasi(self, idx, top_n=10):
        konten = self.df.loc[idx, self.konten_kol]
        kode = self.encoder.transform([konten])
        jarak = cosine_distances(kode, self.bank)
        rek_idx = jarak.argsort()[0, 1:(top_n + 1)]
        return self.df.loc[rek_idx]
        
```
